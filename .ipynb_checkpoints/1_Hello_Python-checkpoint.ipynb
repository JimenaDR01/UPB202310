## Flux Control

```python
points = 85
if points >= 90 and points < 100:
    print("A")
elif points >= 75 and points < 90:
    print("B")
elif points >= 60 and points < 75:
    print("C")
elif points >= 0 and points < 60:
    print("D")
else:
    print("Nel")
```

```python
speed = 60 # km/h
time = input("Cuánto tiempo demoró?: ")
print("Usted recorrió: " + str(speed*float(time)) + " km")
```
